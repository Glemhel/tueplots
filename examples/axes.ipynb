```python
from tueplots import axes
import matplotlib.pyplot as plt
```

```python
axes.lines()
```

```python
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
plt.show()
```

```python
plt.rcParams.update(axes.lines())
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```

```python
plt.rcParams.update(axes.lines(base_width=2.0))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```

```python
plt.rcParams.update(axes.lines(base_width=1., color="red"))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.], label="ABC")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.grid()
plt.show()
```

```python
plt.rcParams.update(axes.lines(facecolor="darkslategray"))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```
