```python
from tueplots import fontsizes, figsize
import matplotlib.pyplot as plt
import matplotlib as mpl
```

Fontsizes are dictionaries that can be passed to `matplotlib.pyplot.rcParams.update()`.

```python
fontsizes.icml()
```

Compare the default font-sizes to, e.g., the ICML style. (To make differences more obvious, we increase the `dpi` value.)

```python
fig, ax = plt.subplots(figsize=figsize.icml(), dpi=150)
ax.plot([1.0, 2.0], [3.0, 4.0])
ax.set_title("Title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
plt.show()
```

```python
plt.rcParams.update(fontsizes.icml())
fig, ax = plt.subplots(figsize=figsize.icml(), dpi=150)
ax.plot([1.0, 2.0], [3.0, 4.0])
ax.set_title("Title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")

plt.show()
```

```python
plt.rcParams.update(fontsizes.neurips())
fig, ax = plt.subplots(figsize=figsize.neurips(), dpi=150)
ax.plot([1.0, 2.0], [3.0, 4.0])
ax.set_title("Title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")

plt.show()
```
