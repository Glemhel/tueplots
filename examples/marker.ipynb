```python
import numpy as np
import matplotlib.pyplot as plt
from tueplots import marker, cycler
from tueplots.constants import markers
from tueplots.constants.color import palettes
```

There are a few marker styles to choose from. (Setup taken from https://matplotlib.org/stable/tutorials/intermediate/color_cycle.html). There are rcParam dictionaries including specific marker styles and full marker cycles.


```python
marker.with_edge()
```

Compare the default options to the specialised markers.  

```python
x = np.linspace(0, np.pi, 25)
offsets = np.linspace(0, 2 * np.pi, 7, endpoint=False)
yy = [np.sin(x + phi) for phi in offsets]
```

```python
plt.subplots(dpi=120)
for y in yy:
    plt.plot(x, y, "o-", linewidth=1)
plt.show()
```

```python
plt.rcParams.update(marker.inverted())

plt.subplots(dpi=120)
for y in yy:
    plt.plot(x, y, "o-", linewidth=1)
plt.show()
```

```python
plt.rcParams.update(marker.with_edge())

plt.subplots(dpi=120)

for y in yy:
    plt.plot(x, y, "o-", linewidth=1)
plt.show()
```

```python
plt.rcParams.update(cycler.cycler(marker=markers.o_sized[:5], color=palettes.pn[:5]))

plt.subplots(dpi=120)

for y in yy:
    plt.plot(x, y, linewidth=1)
plt.show()
```

```python
plt.rcParams.update(marker.with_edge(edgecolor="green", edgewidth=1.0))

plt.subplots(dpi=120)

for y in yy:
    plt.plot(x, y, linewidth=1)
plt.show()
```
