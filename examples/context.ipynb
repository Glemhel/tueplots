```python
from tueplots import context
import matplotlib.pyplot as plt
```

`tueplots` provides a few prepackaged context managers for your plots.

```python
context.neurips2021()
```

Compare the default plots to the context plots below.

```python
fig, ax = plt.subplots(dpi=150)
ax.plot([1., 2.], [3., 4.])
ax.set_title("Title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
plt.show()
```

```python
with context.neurips2021():
    fig, ax = plt.subplots(dpi=150)
    ax.plot([1., 2.], [3., 4.])
    ax.set_title("Title")
    ax.set_xlabel("xlabel")
    ax.set_ylabel("ylabel")
    plt.show()
```

```python
with context.beamer_moml():
    fig, ax = plt.subplots(dpi=150)
    ax.plot([1., 2.], [3., 4.])
    ax.set_title("Title")
    ax.set_xlabel("xlabel")
    ax.set_ylabel("ylabel")
    ax.grid()
    plt.show()
```

```python
with context.beamer_moml_dark_bg():
    fig, ax = plt.subplots(dpi=150)
    ax.plot([1., 2.], [3., 4.])
    ax.set_title("Title")
    ax.set_xlabel("xlabel")
    ax.set_ylabel("ylabel")
    ax.grid()
    plt.show()
```
