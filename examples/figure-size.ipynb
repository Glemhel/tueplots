```python
from tueplots import figsize
import matplotlib.pyplot as plt
```

Figure sizes are tuples. They describe the figure sizes in `inches`, just like what matplotlib expects.

Outputs of `figsize` functions are dictionaries that match `rcParams`. 


```python
icml_size = figsize.icml2022()
icml_size
```

We can use them to make differently sized figures. The height-to-width ratio is (loosely) based on the golden ratio.

```python
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
plt.show()
```

```python
plt.rcParams.update(figsize.icml2022(column="half", nrows=3))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
plt.show()
```

```python
plt.rcParams.update(figsize.neurips2021(nrows=3))

fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
plt.show()
```
