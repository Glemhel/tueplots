```python
from momlplots import figsize
import matplotlib.pyplot as plt
```

Figure sizes are tuples. They describe the figure sizes in `inches`, just like what matplotlib expects.

```python
icml_size = figsize.icml()
icml_size
```

We can use them to make differently sized figures. The height-to-width ratio is (loosely) based on the golden ratio.

```python
fig, ax = plt.subplots(figsize=figsize.icml(column="full"))
ax.plot([1., 2.], [3., 4.])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=figsize.icml(column="half"))
ax.plot([1., 2.], [3., 4.])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=figsize.neurips())
ax.plot([1., 2.], [3., 4.])
plt.show()
```
