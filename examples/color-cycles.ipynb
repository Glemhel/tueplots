```python
import numpy as np
import matplotlib.pyplot as plt 
from tueplots.color import cycler
```

Color cycler define the default color choices in matplotlib (e.g., blue-orange-red-...).
Like most other settings provided in `tueplots`, color cycler are directly compatible with `plt.rcParams.update()`.

To see this, let us generate some lines to be plotted. (Setup taken from https://matplotlib.org/stable/tutorials/intermediate/color_cycle.html).

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)
yy = [np.sin(x + phi) for phi in offsets]
```

The following are the default colors:

```python
for y in yy:
    plt.plot(x, y, linewidth=3)
plt.show()
```

Through the dictionaries provided by, e.g., `color.cycler.probnum()`, we can change the default color behaviour as follows.

```python
plt.rcParams.update(cycler.probnum())
for y in yy:
    plt.plot(x, y, linewidth=3)
plt.show()
```

```python
plt.rcParams.update(cycler.muted())
for y in yy:
    plt.plot(x, y, linewidth=3)
plt.show()
```

```python
plt.rcParams.update(cycler.high_contrast())
for y in yy:
    plt.plot(x, y, linewidth=3)
plt.show()
```

```python
plt.rcParams.update(cycler.bright())
for y in yy:
    plt.plot(x, y, linewidth=3)
plt.show()
```
